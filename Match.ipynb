# Creating Predictors

1. First Predictor: 'venue'.
- need to convert 'home' or 'away' column into a numeric column.
- 'venue indicate whether the team played an 'away' game or 'home'game.
- One of the most important predictor for the outcomes of matches. ( ex: homefield advantage)
- going to convert this to numbers.


```python
# There are two unique values. ('away','home')
# using ".astype" it is converting from'string' into 'category' type, and 
# then using ".cat.codes" will convert 'categories' into the 'integer' type.
# this will create a new column known as 'venue_code'
# It will contain numbers

match["venue_code"] = match["venue"].astype("category").cat.codes
```

```python
match
```

 2. Second Predictor : 'opponent'.
 - Will follow the same method used above.

```python
match["opp_code"] = match["opponent"].astype("category").cat.codes
match
```

```python
#Here we are going to remove the colon and the minutes
# Keep the hour
# will do it with a string replacement 
# using regular expression going to repalce the colon and the minutes with nothing 

match["hour"] = match["time"].str.replace(":.+","",regex = True).astype("int")
```

```python
# Create a new column to represent each day with a number
# ex: sunday : 6, Saturday: 5

match["day_code"] = match ["date"].dt.dayofweek

match
```

# Improving precision with Rolling Averages

- Next we are gonna create more predictors to improve the accuracy of the model. 
- To do this we're going to need to split our 'match' dataframe up by 'team'.Because what we want to do is compute "Rolling Averages" for 'team performance'. 
    - How did this team do in the past few games?
    - How many shots did they have ?
    - How many goals were scored againest them? etc.

So we are gonna create a new dataframe called as 'grouped_matches'

```python
# Create new dataframe which is called as 'grouped_matches'
# this will create one dataframe for every squad in our data 

grouped_matches = match.groupby("team")
# grouped_matches.head()
```

```python
# look at a single group

group = grouped_matches.get_group("Manchester City")
```

```python
group
```

- This will give us a 'single group' from the data and this is just 'Manchester city'. So these are all of the matches that manchester city played in.
- So what we wanna do is compute "rolling avaerages".So what we want to say is if we are 'match week 4' how did 'manchester city' do in the 'previous three match weeks' and then we can pass that into the algorithm and the algorithm can use that information to make better predictions about how they'll do in 'match week 4'.

Ex: if they lost "all three of their games" before "match week four" you might expect them to also "lose match week four" because they haven't been playing very well.

 So we are gonna create a function which named as "rolling_averages".This will do is,
 - take a group in (group)
 - it'll take a set of columns (cols) that we want to compute rolling averages for.
 - and then it will take in a set of new columns that we want to assign the rolling averages to. (new_cols)



```python
#create a function which named as 'rolling averages' 



def rolling_averages(group,cols,new_cols):
    # sort the 'group' by date
    # We want it to be sorted in ascending order of date
    # We want to look at the last three matches a team played and what their performance was like

    group = group.sort_values("date")

    # create a variable calling 'rolling stat
    # what we are gonna do is take a set of columns which will be in this column list 
    # compute the rolling averages for those columns
    # closed ='left' parameter is very important
    rolling_stats = group[cols].rolling(3,closed = 'left').mean()
    
    # assign 'rolling_stats' values back to original dataframe
    group[new_cols] = rolling_stats

    # drop missing values
    group = group.dropna(subset =new_cols)

    return group
```

Column name definitions :
- gf - goals for
- ga - goals against
- sh - shots 
- sot - shots on target
- dist - distance (the distance each shot travelled)
- fk - free kicks
- pk - penalty kicks
- pkatt - penalty kick attempts

```python
# create a list of columns that we want to compute rolling averages
cols = ["gf","ga","sh","sot","dist","fk","pk","pkatt"]

```

```python
# create new column list which is just going to be a format 'string'
new_cols = [f"{c}_rolling" for c in cols]

new_cols
```

Now we can call rolling averages for a single group and our group is manchester city. So this is just for the manchester city games.  

```python
rolling_averages(group,cols,new_cols)

# rolling_averages
```

We can apply this to all of our matches

```python
macth_rolling = match.groupby("team").apply(lambda x: rolling_averages(x,cols,new_cols))
```
